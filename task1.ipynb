```python
###dropping rating columns:
data=data.drop(columns = "rating")
```

```python
data.head()
```

```python
###starting column name with capital letters:
new_column_names=[]
for i in data.columns:
   new_column_names.append(i.capitalize())
    
```

```python
data.columns=new_column_names
```

```python
data.head()
```

```python
data.duplicated().sum()
```

```python
#cleaning duplicates:
data.drop_duplicates(inplace=True)
```

```python
data.head()
```

```python
###checking missing values:
data.isna().sum()
```

```python
###dropping NA values:
data.dropna(inplace=True)
```

```python
data.head()
```

```python
###transforming serial no.:
data["Show_id"].unique()
```

```python
###returning data type og showid to integer
data["Show_id"]= data["Show_id"].apply(lambda x: x.split("s")[1])
```

```python
data.head()
```

```python
###changing pattern of date:
data["Date_added"]= data["Date_added"].apply(lambda x:x.replace("/","-"))
```
